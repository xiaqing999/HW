```python
#----------------------------------------------------------------------------------------------------------
# 1. open the file from the requirement from user input, read into lists and create the dictionaries
#----------------------------------------------------------------------------------------------------------
fname = input('Give the name of the movie reviews file:     ')      # require a user to input the file name

try:                  # in case of error
    fhand = open(fname)        # open the file and read
    lines = fhand.readlines()                 # file reading into a list
    fhand.close()   #read all the lines in the file into a list called lines and close
    reviewers = list()           # defined as the names of reviewrs
    rating = dict()          # defined as movies and the corresponding ratings
    reviewer_rating = dict()            #  defined as each reviewr's name, movies and the corresponding ratings
    dict_all = dict()           # defined as all of the names of reviewrs, movies and the corresponding ratings
    q=0
    for line in lines:
        line.rstrip()                                # delete the blank lines and spaces at the end of each line
        import re
        
        reviewers.extend(re.findall('(^.*?),',line))    #  find the names of reviewers
        sublines = line[line.find(','): ]           # the information of each line except for the the names of reviewers      
        rating = dict(re.findall('(.+?),([0-9.]+)',sublines))   # extract dictionary of name of movies and corresponding ratings
     
        if rating !={ }:   # in case of error of the following computations
            reviewer_rating  = {reviewers[q]: rating}   
        q=q+1
        dict_all.update(reviewer_rating)                          # get all of the names of reviewrs, movies and the corresponding ratings
 
#-----------------------------------------------------------------------------------------------------------------------------------------------------------
# 2. ask for the user's option and then compute the corresponding Euclidean distance, use try &except in case of error
#-----------------------------------------------------------------------------------------------------------------------------------------------------------
    import math
    useinp = input("What do you want to do?\nInput 1 for similarity between two reviewers,\nInput 2 for similarity between one reviewer and all others,\nInput 3 to quit: ")
    ## ask for the user's options of computation
    
    if useinp == "1": 
        reviewer1 = input("Provide reviewer1's name: ")   # ask for the first reviewer's name
        reviewer2 = input("Provide reviewer2's name: ")   # ask for the second reviewer's name
        score=0
        try:   ## in case of error
            for aaa,bbb in dict_all[reviewer1].items():           #  the items of reviewer 1 in the dictionary 
                if aaa in dict_all[reviewer2].keys():           #  make sure the key of reviewer 1 is located in the reviewer 2
                    score1 = float(bbb)                                 # the value of reviewer 1
                    score2 = float(dict_all[reviewer2][aaa])         # the value of reviewer 2
                    score = (score2-score1)**2  +score                     # The Euclidean distance between 2 reviewers
            print('    The similarity score between ',reviewer1,' and ',reviewer2,' is: ',round(math.sqrt(score),4))            
        except:
            print("wrong input, please check1")
    
    elif useinp == "2":
        reviewer3 = input("Provide reviewer's name: ")        # ask for the reviewer's name
        try:
            score=0
            print("\nThe similarity scores are:    ")   # make sure this line is printed only once, so I put it here
            for reviewer in reviewers:
                for aaa,bbb in dict_all[reviewer3].items():
                    if aaa in dict_all[reviewer].keys():   
                        score1 = float(bbb)
                        score2 = float(dict_all[reviewer][aaa])
                        score = score+ (score1-score2)**2              
               
                if reviewer3 != reviewer:                # compute Euclidean distance between one reviewer and all others 
                    print( reviewer3,'      ',reviewer,'      ',round(math.sqrt(score),2))
                   
                score = 0
        except:
            print("wrong input, please check2")       
            
    elif useinp == "3":
        print("bye!")    
    else:
        print("input 1, 2 , or 3")       
except:
    print('wrong input, please check3')

```
