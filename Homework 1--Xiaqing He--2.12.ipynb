```python
### Question 3  ### Finished
###  1. Requests a positive integer from the user
###  2. Use iterative loops to repeatedly run this script.

n=input("Enter a positive integer:")

try:
    n1=int(n)
    if n1>=0:
        if n1>2:
            for m in range(2,n1):
                if (n1%m==0):
                    print("It is a composite")
                    break
                elif (m==n1-1):
                    print("It is a prime")
        elif n1==0 or n1==1:   #### 0 and 1 are neither prime nor composite
            print("It's neither prime nor composite")
        elif n1==2:    #### 2 is the least prime
            print("It's  prime")
    else:
        print("You did not enter a positive integer!")
    
except:  
    print("You did not enter a positive integer!")
```
